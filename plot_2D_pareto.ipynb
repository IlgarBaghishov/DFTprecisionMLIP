```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_results = pd.read_csv('data/2D_pareto/results.csv', index_col=0)
df_results.iloc[:,3:] *= 1000
df = df_results.groupby(list(df_results.columns[:3])).median().reset_index()
df_1 = df[df['Train Precision']==1]
df_2 = df[df['Train Precision']==2]
df_3 = df[df['Train Precision']==3]
df_4 = df[df['Train Precision']==4]
df_5 = df[df['Train Precision']==5]
df_6 = df[df['Train Precision']==6]
print(df_6.columns)
```

```python
plt.figure(figsize=(8, 8))
plt.plot(df_1.iloc[:,-5], df_1.iloc[:,-6], label='prec. 1 tested on 1', linestyle=':', c='r',marker='o',zorder=6)
plt.plot(df_1.iloc[:,-1], df_1.iloc[:,-2], label='prec. 1 tested on 6', linestyle='-', c='r',marker='o',zorder=6)
plt.plot(df_2.iloc[:,-5], df_2.iloc[:,-6], label='prec. 2 tested on 2', linestyle=':', c='gold',marker='o',zorder=5)
plt.plot(df_2.iloc[:,-1], df_2.iloc[:,-2], label='prec. 2 tested on 6', linestyle='-', c='gold',marker='o',zorder=5)
plt.plot(df_3.iloc[:,-5], df_3.iloc[:,-6], label='prec. 3 tested on 3', linestyle=':', c='lime',marker='o',zorder=4)
plt.plot(df_3.iloc[:,-1], df_3.iloc[:,-2], label='prec. 3 tested on 6', linestyle='-', c='lime',marker='o',zorder=4)
plt.plot(df_4.iloc[:,-5], df_4.iloc[:,-6], label='prec. 4 tested on 4', linestyle=':', c='deepskyblue',marker='o',zorder=3)
plt.plot(df_4.iloc[:,-1], df_4.iloc[:,-2], label='prec. 4 tested on 6', linestyle='-', c='deepskyblue',marker='o',zorder=3)
# plt.plot(df_5.iloc[:,-5], df_5.iloc[:,-6], label='prec. 5 tested on 5', linestyle=':', c='b',marker='o',zorder=1)
# plt.plot(df_5.iloc[:,-1], df_5.iloc[:,-2], label='prec. 5 tested on 6', linestyle='-', c='b',marker='o',zorder=1)
# plt.plot(df_6.iloc[:,-5], df_6.iloc[:,-6], label='prec. 6 tested on 6', linestyle=':', c='m',marker='o',zorder=2)
plt.plot(df_6.iloc[:,-1], df_6.iloc[:,-2], label='prec. 6 tested on 6', linestyle='-', c='m',marker='o',zorder=2)
plt.xlabel("Force Testing RMSE, meV/A", fontsize=15)
plt.ylabel("Energy Testing RMSE, meV", fontsize=15)
plt.xticks(fontsize=15)
plt.gca().xaxis.set_minor_formatter(plt.ScalarFormatter())
plt.tick_params(axis='x', which='minor', labelsize=15)
plt.yticks(fontsize=15)
plt.scatter(df_1.iloc[4,-5], df_1.iloc[4,-6], c='r', edgecolors='k', marker='o', s=60, zorder=7)
plt.scatter(df_1.iloc[4,-1], df_1.iloc[4,-2], c='r', edgecolors='k', marker='o', s=60, zorder=7)
plt.scatter(df_2.iloc[5,-5], df_2.iloc[5,-6], c='gold', edgecolors='k', marker='o', s=60, zorder=7)
plt.scatter(df_2.iloc[5,-1], df_2.iloc[5,-2], c='gold', edgecolors='k', marker='o', s=60, zorder=7)
plt.scatter(df_3.iloc[5,-5], df_3.iloc[5,-6], c='lime', edgecolors='k', marker='o', s=60, zorder=7)
plt.scatter(df_3.iloc[5,-1], df_3.iloc[5,-2], c='lime', edgecolors='k', marker='o', s=60, zorder=7)
plt.scatter(df_4.iloc[9,-5], df_4.iloc[9,-6], c='deepskyblue', edgecolors='k', marker='o', s=60, zorder=7)
plt.scatter(df_4.iloc[9,-1], df_4.iloc[9,-2], c='deepskyblue', edgecolors='k', marker='o', s=60, zorder=7)
# plt.scatter(df_5.iloc[9,-5], df_5.iloc[9,-6], c='b', edgecolors='k', marker='o', s=60, zorder=7)
# plt.scatter(df_5.iloc[9,-1], df_5.iloc[9,-2], c='b', edgecolors='k', marker='o', s=60, zorder=7)
# plt.scatter(df_6.iloc[9,-5], df_6.iloc[9,-6], c='m', edgecolors='k', marker='o', s=60, zorder=7)
plt.scatter(df_6.iloc[9,-1], df_6.iloc[9,-2], c='m', edgecolors='k', marker='o', s=60, zorder=7)
plt.xscale('log')
plt.yscale('log')
plt.xlim(100,1000)
plt.ylim(4,1000)
plt.legend(loc='lower right', fontsize=12)
plt.show()
```
